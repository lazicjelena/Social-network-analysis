```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,5])
H = nx.path_graph(10)
G.add_nodes_from(H)
G.add_edges_from([(1,2),(2,3),(5,6)])
G.add_node('spam')
G.add_nodes_from('spam',tip = 'string')
G.add_edge(7,'m',bojaa='ljubicasta')
G.add_edge(4,'p')
G[4]['p']['boja']='plava'
G.add_edges_from([('a',9),(8,9)])
nx.draw(G, with_labels=True,font_weight='bold',pos=nx.spring_layout(G))
plt.show()
```

```python
print('Broj grana: '+str(G.number_of_edges()))
print('Brojcvorova: '+str(G.number_of_nodes()))
print(G.nodes())
print(G.edges())
print(G.degree())
a = G.neighbors(9)
G[4]
print(G.nodes.data())
```
